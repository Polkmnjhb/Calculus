## <font color='red'>1.什麼是費米問題？</font> <br> 　　首先我們來談談費米（費曼）這個人，曾經得過諾貝爾獎的他也被尊稱為「原子學之父」，費米在理論與實驗兩方面都有非常出色的成就，可說是二十世紀最具代表性的天才物理學家，同時他在教育方面也有傑出的表現，教導出許多日後相繼獲得諾貝爾獎的學生。<br> 　　費米在參與美國曼哈頓計劃時曾做過一個驚人之舉，全世界第一次核子試爆時，費米將預先準備好的筆記本撕成碎片灑向空中，紙片由於衝擊波而被吹飛了一段距離，<font color='red'>費米於是從紙片飛行的距離推斷出爆炸威力</font>，之後經由實驗證明，費米推算的威力與實際數值相當接近。 <br> 　　<font color='red'>像此類無足夠資訊，卻又必須推論出答案的問題，我們稱之為「費米問題」</font>，而其推論過程有時候比求得答案的結果還重要，此過程稱之為「費米推定」。

## <font color='red'>2.為什麼要學習費米問題？</font> <br> 　　學習數學最重要的兩樣東西，是<font color='red'>邏輯推理能力</font>與<font color='red'>想像力</font>，而在費米推定的過程中，學生必須擁有天馬行空的想像力來想像問題以得到足夠的資訊，並且利用正確的邏輯推理推斷出真實的答案。<br> 　　在解費米問題時，一般只允許使用加減乘除及次方等簡單的運算得出答案，這是為了訓練如何用簡單的方式推得答案；而且你所推論的答案不需要恰恰好與答案相等，只要你推論出的結果與真實答案相差一個數量級(通常是10倍)，都算正確答案。<br> 　　在費米推定的過程中，你也需要紀錄每一步的想法和推論結果，並在得到答案時，試著想想自己的作法與模型有何不適合的地方吧！
